```python
import numpy as np
photos=np.ones([20,200,200,3])
import os
rootDir = 'photos'
for dirName, subdirList, fileList in os.walk(rootDir):
    print('Found directory: %s' % dirName)
    for fname in fileList:
        print('\t%s' % fname)
        print('Found directory: %s' % dirName)  
        
    
from PIL import Image
resize=(200,200)
for f in os.listdir('photos'):
    if f.endswith('.jpg'):
        i=Image.open(f)
        fn , fext = os.path.splitext(f)
        i.thumbnail((resize))
        i.save('resizeimg/{}_resizeimg{}'.format(fn,fext))

photos=np.array(i)
print(photos)
print(np.shape(i))

   
```
